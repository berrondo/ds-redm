```python
from notebook_importer import *
from base2 import df

import pandas as pd
import numpy as np

from IPython.display import HTML
```

```python
# cria um novo dataframe com demandas e suas sequencias de etapas (status)

dh = pd.DataFrame(columns=['so_paras'])
for d in df.demanda.unique():
    #dp = df.loc[df.demanda==d, ('de', 'para')].values.tolist();   dh.at[d, 'sequencia'] = dp
    p = ''.join(df.loc[df.demanda==d, ('tipo')].values.tolist())
    dh.at[d, 'so_paras'] = p
    #if d == 200: break
```

```python
from itertools import groupby

def f(l): return ''.join([i[0] for i in groupby(l)])

dh['fim'] = dh.so_paras
dh.fim = dh.fim.apply(f)
```

```python
def f2(l):
    if len(l) >=2 and l[-2] == '-':
        return l.split('-')[-1]

dh['encerramento'] = dh.fim
dh.encerramento = dh.encerramento.apply(f2)
```

```python
def f3(l):
    if l and len(l) > 2 and l[-2] == '-':
        return l[:-2]
    if not '-' in l:
        return l

dh['fim3'] = dh.fim
dh.fim3 = dh.fim3.apply(f3)
```

```python
import re
        """I = re.compile('[IsSCx-]')
        if I.match(l): return 'I'"""

def f4(l):
    if l:
        if 'M' in l: return 'M'
        if not (set('D-d') - set(l)): return 'D'
        if not (set('F-c') - set(l)): return 'F'
        if not (set('F-s') - set(l)): return 'F'
        if not (set('F-x') - set(l)): return 'F'
        if not (set('R-') -  set(l)): return 'R'
        if not (set('FA-S') - set(l)): return 'A'
        if not (set('FA-C') - set(l)): return 'A'
        if not (set('I-s') - set(l)): return 'I'
        if not (set('I-S') - set(l)): return 'I'
        if not (set('I-C') - set(l)): return 'I'
        if not (set('Is-x') - set(l)): return 'I'
        if not (set('In-x') - set(l)): return 'I'
        if not (set('I-x') - set(l)): return 'I'
        if not (set('nP-') - set(l)):return 'P'
        if not '-' in l:
            if l in ('FDARPI'): return l
            if not (set('F') - set(l)): return 'F'
            if not (set('In') - set(l)): return 'I'
            if not (set('FA') - set(l)): return 'A'
            if not (set('A') - set(l)): return 'A'
            if not (set('nD') - set(l)): return 'D'
            if not (set('nR') - set(l)): return 'R'
            if not (set('nP') - set(l)): return 'P'

dh['processo'] = dh.fim3
dh.processo = dh.processo.apply(f4)
```

```python
dg = dh[dh.processo.isna()].groupby('fim')
HTML(dg.head().to_html())
```
