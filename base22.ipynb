```python
from notebook_importer import *
from base2 import df

import pandas as pd
import numpy as np

from IPython.display import HTML
```

```python
# cria um novo dataframe com demandas e suas sequencias de etapas (status)

dh = pd.DataFrame(columns=['so_paras'])
for d in df.demanda.unique():
    #dp = df.loc[df.demanda==d, ('de', 'para')].values.tolist();   dh.at[d, 'sequencia'] = dp
    p = ''.join(df.loc[df.demanda==d, ('tipo')].values.tolist())
    dh.at[d, 'so_paras'] = p
    #if d == 200: break
```

```python
from itertools import groupby

def f(l): return ''.join([i[0] for i in groupby(l)])

dh['fim'] = dh.so_paras
dh.fim = dh.fim.apply(f)
```

```python
def f2(l):
    if len(l) >=2 and l[-2] == '-':
        return l.split('-')[-1]

dh['fim2'] = dh.fim
dh.fim2 = dh.fim2.apply(f2)
```

```python
def f3(l):
    if l and len(l) > 2 and l[-2] == '-':
        return l[:-2]
    if not '-' in l:
        return l

dh['fim3'] = dh.fim
dh.fim3 = dh.fim3.apply(f3)
```

```python
def f4(l):
    if l:
        if 'M' in l:
            return 'M'
        if not (set('D-d') - set(l)):
            return 'D'
        if not (set('FOa-c') - set(l)):
            return 'F'
        if not (set('Oa-c') - set(l)):
            return 'F'
        if not (set('Fa-c') - set(l)):
            return 'F'
        if not '-' in l:
            if l in ('FDARPI'):
                return l
            if not (set('FOa') - set(l)):
                return 'F'
            if not (set('Fa') - set(l)):
                return 'F'
            if not (set('a') - set(l)):
                return 'F'
            #if not (set('I') - set(l)):
                #return 'I'
            if not (set('Aa') - set(l)):
                return 'A'

dh['fim4'] = dh.fim3
dh.fim4 = dh.fim4.apply(f4)
```

```python
HTML(dh[dh.fim4.isna()].groupby('fim').head().to_html())
```
